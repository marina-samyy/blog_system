```python
import redis
import json

r = redis.Redis(host="localhost", port=6379, decode_responses=True)


def get_cached_posts(page, size):
    return r.get(f"posts:all:{page}:{size}")


def set_cached_posts(page, size, data):
    r.setex(f"posts:all:{page}:{size}", 60, json.dumps(data))


def get_cached_post(post_id):
    return r.get(f"posts:{post_id}")


def set_cached_post(post_id, data):
    r.setex(f"posts:{post_id}", 60, json.dumps(data))


def invalidate_posts():
    for key in r.scan_iter("posts:all:*"):
        r.delete(key)


def invalidate_post(post_id):
    r.delete(f"posts:{post_id}")
```
