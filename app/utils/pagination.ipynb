```python
def paginate(query, page: int = 1, size: int = 10):
    if page < 1:
        page = 1
    if size < 1:
        size = 10

    offset = (page - 1) * size
    return query.offset(offset).limit(size).all()


def get_pagination_meta(total: int, page: int, size: int):
    return {
        "total": total,
        "page": page,
        "size": size,
        "pages": (total // size) + (1 if total % size else 0)
    }
```
